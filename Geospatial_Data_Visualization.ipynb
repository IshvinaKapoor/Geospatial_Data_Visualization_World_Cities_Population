# Creating an Interactive Display 

```python
from geoplotlib.utils import DataAccessObject
```

```python
#function g
```

```python
g = lambda x: '{}-{}'.format(x['Country'],x['City']).title()
```

```python
#creating a small dataframe df4
df4 = df[:5]
df4
```

```python
#using the g function 
#it displays the country and the city name separately 
print(g(df4))
```

```python
#here we are creating a tooltip which will hover over the map and
#give the city and country name 
da3 = DataAccessObject(df3)
geoplotlib.dot(da3 , f_tooltip = g)
geoplotlib.show()
```

```python
#displaying the screen capture in our jupyter notebook

Image(filename = 'de_gb3.png' , height = n , width= n)
```

# Delaunay Triangulation

```python
#cities with population >= 100K from GB and DE
```

```python
#here df5 is a subset of df3
df5 = df3[(df3.Population >= 1e5)]
df5.shape
```

```python
#it shows the dense areas and connects some series
geoplotlib.delaunay(df5 , cmap = 'hot_r')
geoplotlib.show()
```

```python
#displaying the screen capture in our jupyter notebook

Image(filename = 'delaunay.png' , height = n , width= n)
```
